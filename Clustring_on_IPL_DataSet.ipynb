```python
from pycaret.utils import version
version()
```

```python
from pycaret.datasets import get_data
dataset=get_data('index')
```

```python
ipl_dataset=get_data("ipl")
```

```python
from pycaret.clustering import*
#Kmean Clustering
kmeanparameters=setup(ipl_dataset)
```

```python
kmeanmodel=create_model('kmeans',num_clusters=4)
```

```python
newpredictions=assign_model(kmeanmodel)
newpredictions
```

```python
newpredictions.to_csv("KMeanResult.csv")
print("Result file save sucessfully!!")
```

```python
x=save_model(kmeanmodel,'kmeanmodelFile')
```

```python
newdata=get_data('ipl').iloc[:10]
```

```python
newpredictions=predict_model(kmeanmodel,data=newdata)
newpredictions
```
